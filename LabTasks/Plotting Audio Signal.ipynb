```python
import math,random
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
from scipy.fft import fft, fftfreq
```

Importing audio file

```python
samplerate, data = wavfile.read('applause.wav')
print(samplerate)
```

```python
len(data)
```

```python
duration = len(data)/samplerate
print(duration) #The duration of the audio signal
```

Plotting in Time Domain including both channels

```python
t = np.linspace(0,duration,len(data))
plt.plot(t,data)
plt.xlabel('Time,s')
plt.ylabel('Amplitude')
plt.title('Plotting Audio Signal in Time Domain')
```

Plotting spectrogram to show the amount of energy for particular frequencies at particular times.

```python
powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(data[:,0], Fs=samplerate)
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

Plotting in Frequency Domain

```python

print(data.shape, samplerate)
signal = data[:samplerate,0] #Since audio is stereo; just grab one channel
Signal = fft(signal)
plt.plot(np.abs(Signal[:samplerate//2]), lw=0.15)
plt.xlabel("frequency, Hz")
plt.ylabel("Amplitude, units")
plt.show()
```
