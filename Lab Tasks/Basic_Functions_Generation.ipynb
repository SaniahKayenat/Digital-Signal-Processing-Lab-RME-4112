```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import math,random
```

**Sine Curve**

```python
A = 5
f = 2
t = np.linspace(0,1,100)
phi=0
y = A*np.sin(2*np.pi*f*t)
# print(t)
plt.plot(y)
plt.title('Continuous Analog Signal')
plt.ylabel('Amplitude')
plt.xlabel('Time')
```

**Cosine Curve**

```python
A = 5
f = 2
t = np.linspace(0,1,100)
phi=0
y = A*np.cos(2*np.pi*f*t)
plt.plot(y)
plt.title('Continuous Analog Signal')
plt.ylabel('Amplitude')
plt.xlabel('Time')
```

**A Curve Using Random Values**

```python
t_random = np.linspace(0,1,100)
z = np.random.rand(100)
t=t_random.tolist()

plt.fill_between(t, z,color='g')
plt.plot(t,z)
plt.title('Continuous Random-Time Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

**Unit Step Response**

```python
t_range = np.linspace(-5,11,10)
# print(t_range)
curve=[]
for i in t_range:
  if i >=0:
    curve.append(1)
  else:
    curve.append(0)
plt.stem(t_range,curve)
```

**Unit Ramp Response**

```python
curve = []
for i in t_range:
  if i<0:
    curve.append(0)
  else:
    curve.append(i)
plt.stem(t_range,curve)
```

**Unit Impuse Response**

```python
curve=[]
t=t_range.tolist()
for i in t:
    if i==0:
        curve.append(1)
    else:
        curve.append(0)
if 0 not in t:

  t.append(0)
  curve.append(1)
# print(t_range,curve)
plt.stem(t, curve)
```

**Exponential Curve**

```python
curve = []
for i in t_range:
  curve.append(np.exp(i))
plt.stem(t_range, curve)
```
