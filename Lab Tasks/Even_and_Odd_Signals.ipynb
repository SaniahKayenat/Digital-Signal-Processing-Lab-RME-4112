A signal is even if and only if x(t) = x(-t) or x[n] = x[-n]
while a signal is odd if and only if x(t) = -x(- t) or x[n] = -x[-n].

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

**Defining three functions**

```python
x = np.linspace(0,50,100)
y11 = x**3 + 2*x
y12 = -x**3 + 2*(-x)
y21 = x**4 + 7
y22 = (-x)**4 + 7
y31 = x**2 +3*x - 5
y32 = -x**2 +3*(-x) - 5
```

**Defining a function to identify whether a given signal is Even or Odd**

```python
def identify(x,y1,y2):
  if (y2==y1).all():
    print('The Given Signal is an Even signal')
  elif (y2==(-y1)).all():
    print('The Given Signal is an Odd signal')
  else:
    print('The Given Signal  is Neither Even nor Odd.')


  # print(len(x),len(y1),x.shape,y1.shape)
  figure, axis = plt.subplots(2, 2)
  axis[0, 0].plot(x, y1)
  axis[0, 0].set_title("F(x)")
  axis[0, 1].plot(x, y2)
  axis[0, 1].set_title("F(-x)")
  plt.show()

```

**Calling the function on the predefined functions**

```python
identify(x,y11,y12)
```

```python
identify(x,y21,y22)
```

```python
identify(x,y31,y32)
```
